## Comapring with other market segmentation dataset


```python
def wrangle(path):
    ds1 = pd.read_csv(path, encoding="utf-8")
    print(ds1.info())
    print(ds1.nunique())
    display(ds1.head())
    return ds1
```

```python
count = 1
result_dict = {}
result_name = {}
bold_yellow_text = "\033[1;93m"

reset_style = "\033[0m"
line = "-------------------------||---------------------------"

terminal_width = 80  # Assuming a standard terminal width of 80 columns
text_width = len(line)
spaces_needed = (terminal_width - text_width) // 2

# Create the centered line with spaces on both sides
centered_line = " " * spaces_needed + line

for root, _, files in os.walk("dataset5"):
    for filename in files:
        if ".csv" in filename:
            ds = f"ds{count}"
            print(f"{bold_yellow_text}{ds}{reset_style}")
            print(filename)
            new_path = os.path.join(root, filename)
            result_dict[ds] = wrangle(new_path)
            result_name[ds] = filename
            count += 1
            print(f"{bold_yellow_text}{centered_line}{reset_style}")
           
```

```python
def rename_csv_files(folder):
    for root, _, files in os.walk(folder):
        for filename in files:
            if " " in filename:
                old_path = os.path.join(root, filename)
                new_filename = filename.replace(" ", "_")
                new_path = os.path.join(root, new_filename)
                
                # Check if the new file name already exists, and if so, remove it
                if os.path.exists(new_path):
                    os.remove(new_path)
                
                # Rename the file
                os.rename(old_path, new_path)
```
