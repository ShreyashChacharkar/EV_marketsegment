```python
def wrangle(path):
    ds1 = pd.read_csv(path, encoding="utf-8")
    print(ds1.info())
    print(ds1.nunique())
    display(ds1.head())
    return ds1
```

```python
count = 1
result_dict = {}
result_name = {}
bold_yellow_text = "\033[1;93m"

reset_style = "\033[0m"
line = "-------------------------||---------------------------"

terminal_width = 80  # Assuming a standard terminal width of 80 columns
text_width = len(line)
spaces_needed = (terminal_width - text_width) // 2

# Create the centered line with spaces on both sides
centered_line = " " * spaces_needed + line

for root, _, files in os.walk("dataset5"):
    for filename in files:
        if ".csv" in filename:
            ds = f"ds{count}"
            print(f"{bold_yellow_text}{ds}{reset_style}")
            print(filename)
            new_path = os.path.join(root, filename)
            result_dict[ds] = wrangle(new_path)
            result_name[ds] = filename
            count += 1
            print(f"{bold_yellow_text}{centered_line}{reset_style}")
           
```

```python
def rename_csv_files(folder):
    for root, _, files in os.walk(folder):
        for filename in files:
            if " " in filename:
                old_path = os.path.join(root, filename)
                new_filename = filename.replace(" ", "_")
                new_path = os.path.join(root, new_filename)
                
                # Check if the new file name already exists, and if so, remove it
                if os.path.exists(new_path):
                    os.remove(new_path)
                
                # Rename the file
                os.rename(old_path, new_path)
```

```python
mapped_dict = {"Yes":1,"No":0,"Married":1,"Single":0,"Post Graduate":1,"Graduate":0,"Salaried":1, "Business":0}
ds9["Personal loan"] = ds9["Personal loan"].map(mapped_dict)
ds9["House Loan"] = ds9["House Loan"].map(mapped_dict)
ds9["Marrital Status"] = ds9["Marrital Status"].map(mapped_dict)
ds9["Profession"] = ds9["Profession"].map(mapped_dict)
ds9["Education"] = ds9["Education"].map(mapped_dict)
ds9["Wife Working"] = ds9["Wife Working"].map(mapped_dict)
```

```python
df1 = ds9.drop(columns="Make")
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

k_range = range(2, 10)
inertias = []
k_result = []
silhouette_scores = []
for k in k_range:
    model = KMeans(n_clusters=k, n_init=10, random_state=0)
    model.fit(df1)
    labels = model.predict(df1)
    silhouette_score_ = silhouette_score(df1, labels)
    silhouette_scores.append(silhouette_score_)
    inertias.append(model.inertia_)
    k_result.append(labels)

plt.bar(k_range, inertias)
plt.grid(False)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('sum of within cluster distances')
plt.title('Elbow Method for Optimal k')
plt.xticks(k_range)  # Set x-axis ticks to match k values
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score
ARI_scores = []
for label in k_result:
    boot_n =200
    sample_size =len(label)
    ari_boot = []
    bootstrap_samples = []
    for i in range(200):
        # Draw random indices with replacement
        random_indices = np.random.choice(label, size=sample_size, replace=True)
        # Create a bootstrap sample using the drawn indices
        bootstrap_sample = [label[idx] for idx in random_indices]
        # Append the bootstrap sample to the list
        bootstrap_samples.append(bootstrap_sample)
        ari = adjusted_rand_score(label, bootstrap_sample)
        ari_boot.append(ari+0.01)
    ARI_scores.append(ari_boot)
plt.boxplot(ARI_scores, labels=range(2, 10))
plt.grid(False)
plt.xlabel('Number of segments')
plt.ylabel('Adjusted Rand index')
plt.title('Adjusted Rand Index Distribution')
plt.show()
```

```python
ds19.iat[8,1] = "2 cr"
ds19.iat[12,1] = "1.9 cr"
ds19.iat[22,1] = "1.1 cr"
ds19.iat[25,1] = "1.2 cr"
ds19.drop(columns=["Top_speed","Boot_space"], inplace=True)
ds19.fillna("Automatic")

```

```python
ds19
```

```python
ls = ds19["Drive_range"]
ls1 = []
for item in ls:
    if "km/full charge" in item:
        item = item.replace("km/full charge","")
        item = item.replace(" ","")
        item = item.replace("s","")
        item = item.split("-")
        if len(item) > 1:
            item =round(float(2* float(item[0]) - float(item[1])),2)
        else:
            item = float(item[0])
        ls1.append(item)
    elif "cr" in item:
        item = item.replace("cr","")
        item = item.replace(" ","")
        item = item.split("-")
        if len(item) > 1:
            item =float(2* float(item[0]) - float(item[1]))
        else:
            item = float(item[0])*100
        ls1.append(item)
ds19["Drive_range"] = ls1
```

```python
ls = ds19["Car_price"]
ls1 = []
for item in ls:
    if "lakh" in item:
        item = item.replace("lakh","")
        item = item.replace(" ","")
        item = item.split("-")
        if len(item) > 1:
            item =round(float(2* float(item[0]) - float(item[1])),2)
        else:
            item = float(item[0])
        ls1.append(item)
    elif "cr" in item:
        item = item.replace("cr","")
        item = item.replace(" ","")
        item = item.split("-")
        if len(item) > 1:
            item =float(2* float(item[0]) - float(item[1]))
        else:
            item = float(item[0])*100
        ls1.append(item)
```

```python
ds19["Car_price"] = ls1
ds19["Car_price"] = (ds19["Car_price"]*100_000)
```
